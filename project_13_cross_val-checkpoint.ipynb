```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)          #scaling the dataset
```

```python
# CREATING A FUNCTION FOR FINDING OUT THE BEST RANDOM STATE HAVING THE MAX R2  SCORE 
def maxr2_Score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=r_state,test_size=0.30)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2_score corr to random state',r_state,'is:',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2 score corr to ",final_r_state,'is',max_r_score)       
    return final_r_state 
```

```python
# FINDING  BEST RANDOM STATE FOR LINEAR REG WITH HIGHest R2 SCORE
lreg=LinearRegression()
r_state=maxr2_Score(lreg,x,y)
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=53,test_size=0.30)
    #random state max r2 score corr  is 53

lr=LinearRegression()
lr.fit(x_train,y_train)
ylreg=lr.predict(x_test)
print(lr.score(x_test, y_test)*100)
print(mean_squared_error(ylreg, y_test))
print(mean_absolute_error(ylreg, y_test))
print(np.sqrt(mean_squared_error(ylreg, y_test)))
print(r2_score(y_test,ylreg))
```
