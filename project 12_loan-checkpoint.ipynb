```python
# importing PCA technnique for finding best corr columns
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
x_pca=pca.fit_transform(x)
```

```python
# split data into train and test set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.22)
```

```python
#importing various ensemble methods,boosting methoods, metrics and other models to train test :-
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
## decision tree classifier                # USING GRID SEARCH FOR FINDING BEST PARAMETERS
dtc=DecisionTreeClassifier()
param={'criterion':['gini','entropy']}
clf=GridSearchCV(dtc,param)
clf.fit(x_pca,y)
clf.best_params_
```

```python
## KNEIGHBOURS CALSSIFIER ()
knn=KNeighborsClassifier()
parameters={'n_neighbors':[i+1 for i in range(10)]}
clf=GridSearchCV(knn,parameters)
clf.fit(x_pca,y)
clf.best_params_
```

```python
# random forest calssifier
rfr=RandomForestClassifier()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfr,parameters)
clfr.fit(x_pca,y)
clfr.best_params_
```

```python
# support vector calssifier
svc=SVC()
parameters={'kernel':['rbf','poly','linear']}
clfr=GridSearchCV(svc,parameters ,scoring='accuracy', cv=10)
clfr.fit(x_pca,y)
clfr.best_params_
```

```python
# FINDING THE BEST RANDOM STATE HAVING HIGHEST R2 SCORE :-
# for RANDOM FOREST CLASSIFIER (ENSEMBLE METHOD)
max_acc_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=r,test_size=0.25)
    rfc=RandomForestClassifier(n_estimators=50)
    rfc.fit(x_train,y_train)
    y_rfc=rfc.predict(x_test)
    acc=accuracy_score(y_test,y_rfc)
    print('acc_Score corr to random state',r,'is:',acc)
    if acc>max_acc_score:
        max_acc_score=acc
        final_r_state=r

print("max acc score corr to ",final_r_state,'is',max_acc_score)  
```

```python
# FOR SUPPORT VECTOR CLASSIFIER (SVC)
max_acc_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=r,test_size=0.25)
    svc=SVC(kernel='linear')
    svc.fit(x_train,y_train)
    y_svc=svc.predict(x_test)
    acc=accuracy_score(y_test,y_svc)
    print('acc_Score corr to random state',r,'is:',acc)
    if acc>max_acc_score:
        max_acc_score=acc
        final_r_state=r

print("max acc score corr to ",final_r_state,'is',max_acc_score)  
```

```python
# FOR LOGISTIC REGRESSION(LRG)
max_acc_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=r,test_size=0.25)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    y_lr=lr.predict(x_test)
    acc=accuracy_score(y_test,y_lr)
    print('acc_Score corr to random state',r,'is:',acc)
    if acc>max_acc_score:
        max_acc_score=acc
        final_r_state=r

print("max acc score corr to ",final_r_state,'is',max_acc_score)  
```

```python
# FOR BOOSITNG * GRADIENT BOOSTING CLASSIFIER
max_acc_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=r,test_size=0.25)
    gbc=GradientBoostingClassifier()
    gbc.fit(x_train,y_train)
    y_gbc=gbc.predict(x_test)
    acc=accuracy_score(y_test,y_gbc)
    print('acc_Score corr to random state',r,'is:',acc)
    if acc>max_acc_score:
        max_acc_score=acc
        final_r_state=r

print("max acc score corr to ",final_r_state,'is',max_acc_score)  
```

```python
##****************  TUNINIG THE PARAMETERS******************************
rfc=RandomForestClassifier(n_estimators =50)
knn=KNeighborsClassifier(n_neighbors=9)
svc=SVC(gamma=0.1,kernel='linear')
ada=AdaBoostClassifier()
gbc=GradientBoostingClassifier(learning_rate=0.1,n_estimators=1000)
dtc=DecisionTreeClassifier(criterion="entropy")
LR=LogisticRegression()
```

```python
models = []
models.append(('randomforestclassifier', rfc))
models.append(('kneighborsclassifier', knn))
models.append(('svc', svc))
models.append(('ada boost classifier', ada))
models.append(('gradient boosting classifier', gbc))
models.append(('decisiontreeclassifier', dtc))
models.append(('logistic regression', LR))


```

```python
# import metrics ,scoring parameters and roc curves ......
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,auc,roc_auc_score,roc_curve,f1_score
from sklearn.model_selection import cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=63,test_size=.25) 
# best r state was 63 for svc giving an accuracy score of greater than 87 
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
F1=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("acc_score=",AS)
    score.append(AS*100)
    print('\n')
    f1=f1_score(y_test,pre)
    print("f1_score=",f1)
    F1.append(f1*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('cross_val_sc=',sc)
    cvs.append(sc*100)
    print('\n')
    fpr,tpr,thresholds=roc_curve(y_test,pre)
    roc_auc=auc(fpr,tpr)
    print("roc_auc_sc=",roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('classification_repo\n',classification_report(y_test,pre))
    print('\n')
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(fpr,tpr)
    plt.plot([0,1],[0,1],'r--')
    plt.xlabel('TPR')
    plt.ylabel('FPR')
    print('\n\n')
```

```python
result=pd.DataFrame({'model':Model,'F1-SCORE':F1,'Accu_score':score,'cross_val_Score':cvs,'roc-score':rocscore})
result
```
